```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('24.csv')

# Compute Avg Diameter in KM
df['Avg_Diameter_KM'] = (df['Est Dia in KM(min)'] + df['Est Dia in KM(max)']) / 2

# ================================
# 📊 BAR CHARTS
# ================================

# 1. Hazardous vs Non-Hazardous
plt.figure()
df['Hazardous'].value_counts().plot(kind='bar', color=['green', 'red'])
plt.title('Hazardous vs Non-Hazardous')
plt.xlabel('Hazardous')
plt.ylabel('Count')
plt.xticks(ticks=[0,1], labels=['False', 'True'], rotation=0)
plt.show()

# 2. Absolute Magnitude Distribution (binned)
plt.figure()
df['Absolute Magnitude'].plot.hist(bins=10, color='purple', edgecolor='black')
plt.title('Distribution of Absolute Magnitude')
plt.xlabel('Absolute Magnitude')
plt.ylabel('Count')
plt.show()

# 3. Size category (KM) bins
plt.figure()
df['Size Category'] = pd.cut(df['Avg_Diameter_KM'], bins=[0, 0.1, 0.5, 1, 2], labels=['Tiny', 'Small', 'Medium', 'Large'])
df['Size Category'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Asteroid Size Categories (by Avg Diameter in KM)')
plt.xlabel('Size Category')
plt.ylabel('Count')
plt.show()

# ================================
# 🔴 SCATTER PLOTS
# ================================

# Helper: General scatter function
def scatter(x, y, title, xlabel, ylabel):
    plt.figure()
    sns.scatterplot(data=df, x=x, y=y, hue='Hazardous')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend(title='Hazardous')
    plt.show()

# 1. Avg Diameter vs Orbital Period
scatter('Avg_Diameter_KM', 'Orbital Period', 'Diameter vs Orbital Period', 'Avg Diameter (km)', 'Orbital Period (days)')

# 2. Max Diameter vs Mean Motion
scatter('Est Dia in KM(max)', 'Mean Motion', 'Max Diameter vs Mean Motion', 'Max Diameter (km)', 'Mean Motion')

# 3. Max Diameter (M) vs Perihelion Distance
scatter('Est Dia in M(max)', 'Perihelion Distance', 'Max Diameter vs Perihelion Distance', 'Max Diameter (m)', 'Perihelion Distance')

# 4. Mean Anomaly vs Aphelion Distance
scatter('Mean Anomaly', 'Aphelion Dist', 'Mean Anomaly vs Aphelion Distance', 'Mean Anomaly (°)', 'Aphelion Distance (AU)')

# 5. Absolute Magnitude vs Avg Diameter
scatter('Absolute Magnitude', 'Avg_Diameter_KM', 'Absolute Magnitude vs Diameter', 'Absolute Magnitude', 'Avg Diameter (km)')

# 6. Perihelion Arg vs Perihelion Time
scatter('Perihelion Arg', 'Perihelion Time', 'Perihelion Argument vs Perihelion Time', 'Perihelion Arg (°)', 'Perihelion Time')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('final_train.csv')

# Compute Avg Diameter in KM, assuming we only have 'Est Dia in Feet(max)'
# Convert feet to km: 1 foot = 0.0003048 km
df['Avg_Diameter_KM'] = df['Est Dia in Feet(max)'] * 0.0003048

# ================================
# 📊 BAR CHARTS
# ================================

# 1. Hazardous vs Non-Hazardous
plt.figure()
df['Hazardous'].value_counts().plot(kind='bar', color=['green', 'red'])
plt.title('Hazardous vs Non-Hazardous')
plt.xlabel('Hazardous')
plt.ylabel('Count')
plt.xticks(ticks=[0,1], labels=['False', 'True'], rotation=0)
plt.show()

# 2. Absolute Magnitude Distribution (binned)
plt.figure()
df['Absolute Magnitude'].plot.hist(bins=10, color='purple', edgecolor='black')
plt.title('Distribution of Absolute Magnitude')
plt.xlabel('Absolute Magnitude')
plt.ylabel('Count')
plt.show()

# 3. Size category (KM) bins
plt.figure()
df['Size Category'] = pd.cut(df['Avg_Diameter_KM'], bins=[0, 0.1, 0.5, 1, 2], labels=['Tiny', 'Small', 'Medium', 'Large'])
df['Size Category'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Asteroid Size Categories (by Avg Diameter in KM)')
plt.xlabel('Size Category')
plt.ylabel('Count')
plt.show()

# ================================
# 🔴 SCATTER PLOTS
# ================================

# Helper: General scatter function
def scatter(x, y, title, xlabel, ylabel):
    plt.figure()
    sns.scatterplot(data=df, x=x, y=y, hue='Hazardous')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend(title='Hazardous')
    plt.show()

# 1. Avg Diameter vs Orbital Period
scatter('Avg_Diameter_KM', 'Orbital Period', 'Diameter vs Orbital Period', 'Avg Diameter (km)', 'Orbital Period (days)')

# 2. Max Diameter vs Mean Motion
# Since we only have 'Est Dia in Feet(max)', we'll use the derived 'Avg_Diameter_KM' and convert back to feet if needed
scatter('Avg_Diameter_KM', 'Mean Motion', 'Max Diameter vs Mean Motion', 'Max Diameter (km)', 'Mean Motion') # Using Avg_Diameter_KM as a proxy

# 3. Max Diameter (M) vs Perihelion Distance
# Assuming 'Est Dia in Feet(max)' represents max diameter in feet, convert to meters (1 foot = 0.3048 meters)
df['Est Dia in M(max)'] = df['Est Dia in Feet(max)'] * 0.3048
scatter('Est Dia in M(max)', 'Perihelion Distance', 'Max Diameter vs Perihelion Distance', 'Max Diameter (m)', 'Perihelion Distance')

# 4. Mean Anomaly vs Aphelion Distance
scatter('Mean Anomaly', 'Aphelion Dist', 'Mean Anomaly vs Aphelion Distance', 'Mean Anomaly (°)', 'Aphelion Distance (AU)')

# 5. Absolute Magnitude vs Avg Diameter
scatter('Absolute Magnitude', 'Avg_Diameter_KM', 'Absolute Magnitude vs Diameter', 'Absolute Magnitude', 'Avg Diameter (km)')

# 6. Perihelion Arg vs Perihelion Time
scatter('Perihelion Arg', 'Perihelion Time', 'Perihelion Argument vs Perihelion Time', 'Perihelion Arg (°)', 'Perihelion Time')

# 4. Mean Anomaly vs Aphelion Distance
scatter('Mean Anomaly', 'Aphelion Dist', 'Mean Anomaly vs Aphelion Distance', 'Mean Anomaly (°)', 'Aphelion Distance (AU)')

# 5. Absolute Magnitude vs Avg Diameter
scatter('Absolute Magnitude', 'Avg_Diameter_KM', 'Absolute Magnitude vs Diameter', 'Absolute Magnitude', 'Avg Diameter (km)')

# 6. Perihelion Arg vs Perihelion Time
scatter('Perihelion Arg', 'Perihelion Time', 'Perihelion Argument vs Perihelion Time', 'Perihelion Arg (°)', 'Perihelion Time')
```
